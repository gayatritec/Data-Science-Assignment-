#Estimation And Confidence Intervals

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import t
import matplotlib.pyplot as plt
```

**Background** :

In quality control processes, especially when dealing with high-value items, destructive sampling is a necessary but costly method to ensure product quality. The test to determine whether an item meets the quality standards destroys the item, leading to the requirement of small sample sizes due to cost constraints.

**Scenario** :

A manufacturer of print-heads for personal computers is interested in estimating the mean durability of their print-heads in terms of the number of characters printed before failure. To assess this, the manufacturer conducts a study on a small sample of print-heads due to the destructive nature of the testing process.

**Data**:

A total of 15 print-heads were randomly selected and tested until failure. The durability of each print-head (in millions of characters) was recorded as follows:
1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29




```python
data =np.array([1.33, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48,1.20, 1.33, 1.18, 1.22, 1.29])
data
```

```python
print("Mean:",data.mean()) #sample mean of the data
print("Standard Deviation:",data.std())#sample standard deviation
n=len(data)#sample size
print("Sample Size:",n)#print sample size
```

# Build 99% Confidence Interval Using Sample Standard Deviation:

#t-distribution

In the question standard deviation of population is not mension,so here i am using t-distribution.
* confidence level is 99% and the degree of freedom is 14.
* S0 t distribution value is 2.624 from T-table.

```python
Confidence_level=0.99 #Confidence level
df=n-1 #degree of freedom
confidence_interval_UL=data.mean()+2.624*data.std()/np.sqrt(n)#confidence interval for upper range
round(confidence_interval_UL,2)#Rounding the data into 2 decimals
```

```python
confidence_interval_LL=data.mean()-2.624*data.std()/np.sqrt(n)#confidence interval of lower range
round(confidence_interval_LL,2)#Rounding the data into 2 decimal.

```

```python
import seaborn as sns
sns.set(style="darkgrid")
sns.kdeplot(data,color="purple",linestyle="-",fill=True,alpha=0.3,linewidth=2)
plt.axvline(data.mean(), color='black', linestyle='--', linewidth=2, label='Mean')
plt.text(data.mean() + 0.1, 0.02, f'Mean: {data.mean():.2f}', color='black', fontsize=15)
plt.show()
plt.title('Density Curves for data')#adding the title
plt.xlabel('Values')#adding the level of x-axis
plt.ylabel('Density')#adding the level for y-axis
plt.axvspan(confidence_interval_LL, confidence_interval_UL, color='green', alpha=0.2, label='Confidence Interval')#showing confidence interval
plt.legend()#adding legend
plt.show()# Display the plot

```

**Confindence intervals are 1.11 & 1.39, So 99% of the data lies between range of 1.11 to 1.39.**
